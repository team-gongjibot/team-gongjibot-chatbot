```python
def summarize_notice_llm(notice_text: str) -> str:
    system_prompt = """
LLM에 넣기 좋게 다음 텍스트를 요약해봐
"""

    response = client.chat.completions.create(
        model="local-model",  # LM Studio에서 실행 중인 모델 (이름은 무시됨)
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": notice_text}
        ],
        temperature=0.5,
        max_tokens=1024,
        frequency_penalty=0.4,
        presence_penalty=0.1,
    )

    return response.choices[0].message.content.strip()
```

```python
df["summary"] = None
for i in range(20):
    # 테스트용 공지사항 텍스트
    notice_text = df["text"][i]

    # 요약 결과 출력
    result = summarize_notice_llm(notice_text)
    df.loc[i,"summary"] = result
    print(f"{i} : {result}\n\n\n\n\n ")
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("intfloat/multilingual-e5-base")
```

```python
import faiss
import numpy as np

df["embedding"] = df["summary"].apply(lambda x: model.encode(x).tolist())
df["tokenize_doc"] = df["summary"].apply(lambda x: x.split())

dimension = len(df["embedding"][0])
index = faiss.IndexFlatL2(dimension)
doc_embeddings = np.stack(df["embedding"].values)
index.add(doc_embeddings)
```

```python
query = "개교기념일로 인한 보강은 언제야?"
query_embedding = model.encode([query])

top_k = 5
D, I = index.search(query_embedding, len(df))
# distances: np.ndarray (shape: [n_queries, k])
# indices: np.ndarray (shape: [n_queries, k])

for i in range(top_k):
    print(f"{i+1}위 문서 : {df.loc[I[0][i], 'summary']} \n (거리 : {D[0][i]:.4f}) (인덱스 : {I[0][i]})")
```
