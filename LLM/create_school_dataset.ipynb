```python
# ✅ OpenAI 최신 버전 방식으로 설정
from openai import OpenAI

# LM Studio API 서버 연결
client = OpenAI(
    base_url="http://localhost:1234/v1",  # LM Studio 주소
    api_key="lm-studio"  # 아무 문자열도 OK
)
```

```python
def summarize_notice_llm(notice_text: str) -> str:
    system_prompt = """
    Do NOT output <think> or any reasoning trace.  
Return ONLY the final answer that the user can see.
<think></think>부분은 절대 LLM답변에 추가시키지마.
너는 주어진 문서에서 질문-답변(QA) 쌍을 추출하고 생성하는 AI 챗봇이야.
다음 규칙을 엄격하게 지켜야 해:
1. 문서 내용만을 바탕으로 QA를 생성해야 해.
2. 질문과 답변은 문서 내의 정보에 기반해야 하며, 외부 정보를 추가하면 안 돼.
3. 중복된 질문이나 답변은 절대 허용되지 않아.
4. 답변은 자연스럽게, 사람이 답하는 것처럼 해줘.
5. 생성된 모든 QA 쌍은 JSON 형식으로 출력해야 해.
. 영어와 한국어 외의 다른 문자는 포함되면 안 돼.
. 너가 만든걸 데이터셋으로 학습을 시킬거기 때문에 qas가 들어가야해. 
. 한 페이지당 최소 3개에서 최대 10개의 QA 쌍을 생성해야 해.
. 출력은 </think> 토큰을 제외하고 json내용만 출력하도록해.

. JSON 배열 내의 각 객체는 {"question": "질문 내용", "answer": "답변 내용"} 형식을 따라야 해.

질문: 브리트니 스피어스는 어디에서 태어났나요?

<생각> 이 가수의 출생지를 알기 위해 배경 정보를 검색해야 해요. </생각>
<검색> 브리트니 스피어스 출생지 </검색>
<정보> 브리트니 스피어스는 미시시피주 맥콤에서 태어났습니다. </정보>
<생각> 충분한 정보가 확보됐어요. </생각>
<답변> 미시시피주 맥콤 </답변>

"""

    response = client.chat.completions.create(
        model="local-model",  # LM Studio에서 실행 중인 모델 (이름은 무시됨)
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": notice_text}
        ],
        temperature=0.6,
        max_tokens=2048,
        frequency_penalty=0.4,
        presence_penalty=0.1,
    )

    return response.choices[0].message.content.strip()
```

```python
import re

data=[]

for i in range(31,(len(df))):
    # 테스트용 공지사항 텍스트
    notice_text = df["summary"][i]
    context = notice_text
    # 요약 결과 출력
    result = summarize_notice_llm(notice_text)
    print(f"{i} : {result}")
    result = re.sub(r'<think>.*?</think>', '', result, flags=re.DOTALL)
    try:
        qa_list = json.loads(result)
    except:
        continue
    structure_dataset= {
        "context" : context,
        "qas": qa_list
    }
    
    data.append(structure_dataset)
    
data = {
    "data" : data
}

with open("merged_data1.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)
```

```python
import json

with open("merged_data2.json", "r", encoding="utf-8") as f:
    data = json.load(f)

converted_data = []

for entry in data['data']:
    context = entry["context"]
        # 현재 문단의 'qas' (질문-답변 쌍) 순회
    for qa in entry['qas']:
        print(qa)
        question = qa['question'] # 질문 텍스트
            
        if qa['answer']:
                # 여러 답변이 있을 수 있지만, 보통 첫 번째 답변을 사용합니다.
            answer = qa['answer']

                # Qwen3-4B 모델의 입력 형식 (메시지 리스트) 생성
            messages = [
                # 시스템 메시지는 필요에 따라 추가하거나 변경할 수 있습니다.
                {"role": "system", "content": "You are a helpful assistant that answers questions based on the provided context."},
                # 사용자 메시지: 맥락과 질문 포함
                {"role": "user", "content": f"맥락: {context}\n질문: {question}"},
                # 어시스턴트 메시지: 정답 답변
                {"role": "assistant", "content": answer}
            ]

            converted_data.append({"messages": messages})
        else:
            # 답변이 없는 질문을 처리하려면 이 부분을 수정해야 합니다.
            # 예: {"role": "assistant", "content": "답변을 찾을 수 없습니다."}
            # 또는 해당 예제를 스킵. 기본적으로는 스킵합니다.
            pass

with open("collage_dataset.json", 'w', encoding='utf-8') as f:
    # JSON 배열 형태로 저장, ensure_ascii=False로 한글 깨짐 방지
    json.dump(converted_data, f, indent=4, ensure_ascii=False)

print(f"SQuAD 데이터 변환 완료. 총 {len(converted_data)}개의 예제가 저장되었습니다.")
```
